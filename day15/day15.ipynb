```python
# Q1
board = []
moves = []
with open("input.txt") as f:
    while (line := f.readline()) != "\n":
        board.append(list(line.strip()))
        for b in range(len(board[-1])):
            if board[-1][b] == "@":
                robot_y, robot_x = len(board) - 1, b
    while line := f.readline():
        moves += line.strip()

move_coords = {"^": (-1, 0), "v": (1, 0), "<": (0, -1), ">": (0, 1)}

for m in moves:
    dy, dx = move_coords[m]
    temp_y, temp_x = robot_y + dy, robot_x + dx
    while board[temp_y][temp_x] != "#" and board[temp_y][temp_x] != ".":
        temp_y += dy
        temp_x += dx
    if board[temp_y][temp_x] == ".":
        while (temp_y, temp_x) != (robot_y, robot_x):
            board[temp_y][temp_x], board[temp_y - dy][temp_x - dx] = board[temp_y - dy][temp_x - dx], board[temp_y][temp_x]
            temp_y -= dy
            temp_x -= dx
        robot_y, robot_x = robot_y + dy, robot_x + dx
gps_coords = 0
for r in range(len(board)):
    for c in range(len(board[r])):
        if board[r][c] == "O":
            gps_coords += r * 100 + c
print(gps_coords)
```

```python
# Q2
board = []
moves = []
with open("input.txt") as f:
    while (line := f.readline()) != "\n":
        board.append(list(line.strip().replace("#", "##").replace("O", "[]").replace(".", "..").replace("@", "@.")))
        for b in range(len(board[-1])):
            if board[-1][b] == "@":
                robot_y, robot_x = len(board) - 1, b
    while line := f.readline():
        moves += line.strip()

move_coords = {"^": (-1, 0), "v": (1, 0), "<": (0, -1), ">": (0, 1)}
for m in moves:
    dy, dx = move_coords[m]
    if m in ["<", ">"]:
        temp_y, temp_x = robot_y + dy, robot_x + dx
        while board[temp_y][temp_x] != "#" and board[temp_y][temp_x] != ".":
            temp_y += dy
            temp_x += dx
        if board[temp_y][temp_x] == ".":
            while (temp_y, temp_x) != (robot_y, robot_x):
                board[temp_y][temp_x], board[temp_y - dy][temp_x - dx] = board[temp_y - dy][temp_x - dx], board[temp_y][temp_x]
                temp_y -= dy
                temp_x -= dx
            robot_y, robot_x = robot_y + dy, robot_x + dx
    else:
        is_wall = False
        path = []
        curr_row = set([(robot_y, robot_x)])
        next_row = set()
        while curr_row and not is_wall:
            for r, c in list(curr_row):
                temp_y, temp_x = r + dy, c + dx
                if board[temp_y][temp_x] == "#":
                    is_wall = True
                    break
                elif board[temp_y][temp_x] == "[":
                    next_row.add((temp_y, temp_x))
                    next_row.add((temp_y, temp_x + 1))
                elif board[temp_y][temp_x] == "]":
                    next_row.add((temp_y, temp_x))
                    next_row.add((temp_y, temp_x - 1))
            path.append(curr_row)
            curr_row = next_row
            next_row = set()
        if not is_wall:
            for p in path[::-1]:
                for r, c in p:
                    board[r][c], board[r + dy][c + dx] = board[r + dy][c + dx], board[r][c]
            robot_y, robot_x = robot_y + dy, robot_x + dx
print("".join(["".join(b) + "\n" for b in board]))
gps_coords = 0
for r in range(len(board)):
    for c in range(len(board[r])):
        if board[r][c] == "[":
            gps_coords += r * 100 + c
print(gps_coords)
```
